# Exploring the 'brambles' dataset

This notebook uses the 'brambles' dataset from `pydataset` to explore customizations to `seaborn` scatterplots.  The adjustments include changes to the background style and color palette, as well as adjusting the positioning of the legend, and editing the default labels. This notebook also utilizes both the `size` and `hue` parameters to convey the age of each bramble along with its position.

```python
from pydataset import data

df = data('brambles')
# display(data('brambles', show_doc = True))
display(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
sns.set(rc={'axes.facecolor':'#B39880'})
p = sns.color_palette('YlGn', 3)
g = sns.scatterplot(x='x', y='y', data = df, size='age', hue='age', palette=p)
plt.axis('equal')
plt.title('Bramble location and age')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title = 'age')
g.set(xlabel=None, ylabel=None)
plt.show()
```
